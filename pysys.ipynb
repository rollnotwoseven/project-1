```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
data = yf.download('GOOG')

data.head()
```

```python
data['RowNumber'] = range(1,len(data)+1)
data.head()
```

```python
data= data.drop(columns=['Volume'])

data.head()
```

```python
data=data[['RowNumber','Open','High','Low','Close']]

data.head()
```

```python
data['9-day'] = data['Close'].rolling(9).mean()

data.head()
```

```python
data['21-day'] = data['Close'].rolling(21).mean()

data.head()
```

```python
data = data.drop(columns=['(9-day','(21-day'])
```

```python
data.head()
```

```python
data = data.dropna()

data.head()
```

```python
data['RowNumber'] = range(1,len(data)+1)
```

```python
data.head()
```

```python
data['sno'] = data['RowNumber']
```

```python
data.head()
```

```python
data=data.drop(columns=['RowNumber'])
```

```python
data=data[['sno','Open','High','Low','Close','9-day','21-day']]

data.head()
```

```python
data.head()
```

```python
data['9-day'] = data['Close'].rolling(9).mean()
data['21-day'] = data['Close'].rolling(21).mean()

data.head()
```

```python
data = data.dropna()
data.head()
```

```python
data['Sno'] = range(1,len(data)+1)
```

```python
data.head()
```

```python
data = data[['sno','Open','High','Low','Close','9-day','21-day']]

data.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.index,data['Close'],label='GOOG close price')
plt.title('Google (GOOG) close price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
